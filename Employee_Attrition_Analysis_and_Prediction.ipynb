```python
#perform encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Attrition'] = le.fit_transform(data['Attrition'])
data['BusinessTravel'] = le.fit_transform(data['BusinessTravel'])
data['Department'] = le.fit_transform(data['Department'])
data['EducationField'] = le.fit_transform(data['EducationField'])
data['Gender'] = le.fit_transform(data['Gender'])
data['JobRole'] = le.fit_transform(data['JobRole'])
data['MaritalStatus'] = le.fit_transform(data['MaritalStatus'])
data['OverTime'] = le.fit_transform(data['OverTime'])
data['Over18'] = le.fit_transform(data['Over18'])
```

```python
# Calculate the correlation of all columns with the target column
correlations = data.corrwith(data['Attrition'])
print(type(correlations))
# Print the correlation values
print(correlations.sort_values(ascending=True))
```

```python
#remove 3 columns : 'EmployeeCount','Over18','StandardHours'
x = data.drop(['EmployeeCount','Over18','StandardHours','Attrition'],axis=1)
y = data['Attrition']
```

```python
x.shape
```

```python
y.shape
```

```python
#sns.pairplot(data)
```

**Splitting Data into Training and Testing Data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

**Transforming Data using Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
```

**Predicting x_test on Trainined Dataset**

```python
y_pred = lr_model.predict(x_test)
```

**Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

**Accuracy Score**

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

**Classification Report**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
y_pred_prob  = lr_model.predict_proba(x_test)
y_pred_prob
formatted_probabilities = [[format(prob, '.2f') for prob in inner_list] for inner_list in y_pred_prob]

print(formatted_probabilities)
```

**K Fold Cross Validation**

```python
from sklearn.model_selection import cross_val_score,KFold

kfold = KFold(n_splits=10, shuffle=True, random_state=42)
kf_cv_scores = cross_val_score(lr_model, x_train, y_train, cv=kfold ,scoring='accuracy')
print("Scores of All folds : ",kf_cv_scores)
print("K-fold CV average score: %.2f" % kf_cv_scores.mean())
```

**Fitting Data on K Fold Model**

```python
lr_model.fit(x_train,y_train)
y_pred_kfold = lr_model.predict(x_test)
y_pred_kfold
```

```python
from sklearn.metrics import confusion_matrix
kfold_cm = confusion_matrix(y_test,y_pred_kfold)
kfold_cm
```

**Ensemble Learning Using Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(criterion='entropy')
rf_model.fit(x_train,y_train)
```

```python
y_pred_rf = rf_model.predict(x_test)
y_pred_rf
```

```python
from sklearn.metrics import confusion_matrix
rf_cm = confusion_matrix(y_test,y_pred_rf)
rf_cm
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
kf_cv_scores = cross_val_score(rf_model, x_train, y_train, cv=kfold ,scoring='accuracy')
print("Scores of All folds : ",kf_cv_scores)
print("K-fold CV average score: %.2f" % kf_cv_scores.mean())
```

```python
rf_model.fit(x_train,y_train)
y_pred_kfold_rf = rf_model.predict(x_test)
y_pred_kfold_rf
```
